**Load Dataset**

```python
dataset = pd.read_csv('salary.csv')
```

**Summarise Dataset**

```python
print(dataset.shape)
print(dataset.head())
print(dataset.isnull().sum())
```

```python
income_set = set(dataset['income'])
dataset['income'] = dataset['income'].map({'<=50K': 0, '>50K': 1}).astype(int)
print(dataset.head)
```

**Segregate Dataset into X(independent) and Y(dependent)**

```python
X = dataset.iloc[:,:-1].values
X
```

```python
Y = dataset.iloc[:,-1].values
Y
```

**Spliting Data**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.25, random_state = 0)
```

**Standardization**

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

**Finding the Best K-Values**

```python
error = []
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt

#calculating error for k values between 1 and 40
for i in range(1,40):
  model = KNeighborsClassifier(n_neighbors=i)
  model.fit(X_train,Y_train)
  pred_i = model.predict(X_test)
  error.append(np.mean(pred_i!= Y_test))

plt.figure(figsize = (12,6))
plt.plot(range(1,40),error,color='black', linestyle = 'dashed', marker = 'o', markerfacecolor = 'red', markersize = 10)
plt.title('Error Rate K Value')
plt.xlabel('mean error')
plt.ylabel('K Value')
```

**Training**

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors = 30, metric ='minkowski', p =2)
model.fit(X_train, Y_train)
```

**Predicting, whether new customer with Age & Salary will Buy or Not**

```python
age = int(input("Enter New Employee's Age: "))
edu = int(input("Enter New Employee's Education: "))
cg = int(input("Enter New Employee's Capital Gain: "))
wh = int(input("Enter New Employee's Hour's Per week: "))
newEmp = [[age,edu,cg,wh]]
result = model.predict(sc.transform(newEmp))
print(result)

if result == 1:
  print("Emplyee might got Salary above 50k ")
else:
  print("Customer might not got salary above 50k ")
```

**Prediction for all Test Data**

```python
y_pred = model.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred), 1), Y_test.reshape(len(Y_test),1)),1))
```

**Evaluating Model - CONFUSION MATRIX**


```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(Y_test, y_pred)
print("Confusion Matrix: ")
print(cm)
print("Accuracy of the Model: {0}%".format(accuracy_score(Y_test,y_pred)*100))
```
