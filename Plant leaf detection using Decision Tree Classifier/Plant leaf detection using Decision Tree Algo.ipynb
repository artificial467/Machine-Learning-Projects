Import Basic Libraries

```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
```

**Load** **Dataset**

```python
dataset = load_iris()
```

**Summarize Dataset **

```python
print(dataset.data)
print("\n")
print(dataset.target)
print("\n")

print(dataset.data.shape)
```

Segregate Dataset into X(independent ) & Y (dependent)

```python
X = pd.DataFrame(dataset.data,columns=dataset.feature_names )
X
```

```python
Y = dataset.target
Y
```

**Splitting Dataset into Train & Test**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = 0)
print(X_train.shape)
print(X_test.shape)
```

**Finding the best Max_Depth**

```python
accuracy = []
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
for i in range(1,10):
  model = DecisionTreeClassifier(max_depth=i, random_state = 0)
  model.fit(X_train, Y_train)
  pred = model.predict(X_test)
  score = accuracy_score(Y_test, pred)
  accuracy.append(score)

plt.figure(figsize=(12,6))
plt.plot(range(1,10), accuracy, color = 'red', linestyle = 'dashed', marker = 'o', mec = 'blue', mfc = 'hotpink', markersize = 20, linewidth = 5)
plt.xlabel('predicted value')
plt.ylabel('score value')
plt.title('Graph betwenn pred and score')
plt.grid()
```

Training

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(criterion = 'entropy', max_depth=3, random_state = 0)
model.fit(X_train,Y_train)
```

**Prediction**

```python
y_pred = model.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), Y_test.reshape(len(Y_test),1)),1))
```

**Accuracy Score**

```python
from sklearn.metrics import accuracy_score
print("Accuracy score of model : {0}%".format(accuracy_score(Y_test,y_pred)*100))
```
