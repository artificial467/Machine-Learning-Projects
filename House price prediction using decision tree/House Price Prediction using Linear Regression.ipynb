**Summarize Data**

**Visualize Data**

```python
plt.scatter(dataset.area, dataset.price, c='g', linestyle='--',marker = 'o' )
plt.xlabel('AREA')
plt.ylabel("PRICE")

plt.title('Graph between Area and price')
```

**Segregate Dataset into X & Y**

```python
X = dataset.drop('price', axis = 'columns')
X
```

```python
Y = dataset.price
```

```python
Y
```

**Training Dataset using Linear Regression**

```python
model = LinearRegression()
model.fit(X,Y)
```

**Predicted Price for Land Sq.feet of 1500**

```python
x = 6660
LandAreainSqFt = [[x]]
PredictedmodelResult = model.predict(LandAreainSqFt)
print(PredictedmodelResult)
```

**Theory Calculation**


Y = m*X + C (m is coefficient and b is intercept)

```python
m = model.coef_
print(m)
```

intercept -b

```python
b = model.intercept_
print(b)
```

y = m*x + C


```python
y = m*x + b
print("the price of {0} Square feet Land is : {1}".format(X,y[0]))
```
